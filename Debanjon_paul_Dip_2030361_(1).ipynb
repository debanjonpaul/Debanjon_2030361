Problem 03:Find the mean of the 3rd column of the matrix without using the built-in function and loop.

```python
import numpy as np

# Step 1: Generate 10x5 random matrix
random_matrix = np.random.rand(10, 5)
print("Random Matrix:\n", random_matrix)

# Step 2: Select 3rd column (index 2)
third_col = random_matrix[:, 2]

# Step 3: Find mean manually (no np.mean, no loop)
mean_third_col = np.sum(third_col) / len(third_col)

# Step 4: Display result
print("\n3rd Column:", third_col)
print("Mean of 3rd Column:", mean_third_col)
```

Problem 04:Generate the vector [25, 16, 9, 4, 1] and then find the square root of each element of the vector using for loop. Hint: Use (np.sqrt())
function to find the square root of a number, don't use any other built-in function.



```python
import numpy as np

# Step 1: Create the given vector
vector = np.array([25, 16, 9, 4, 1])
print("Original Vector:", vector)

# Step 2: Create an empty list to store results
sqrt_vector = []

# Step 3: Use for loop to find square root of each element
for val in vector:
    sqrt_val = np.sqrt(val)
    sqrt_vector.append(sqrt_val)

# Step 4: Convert list to numpy array (optional)
sqrt_vector = np.array(sqrt_vector)

# Step 5: Display result
print("Square Root of each element:", sqrt_vector)
```

Problem 05:
Write a function that takes a number and returns a vector like the previous problem. For example, if the input is 5, then the output should be `[25, 16, 9, 4, 1]`. If the input is 10, then the output should be `[100, 81, 64, 49, 36, 25, 16, 9, 4, 1]`.

```python
import numpy as np

# Step 1: Define the function
def create_square_vector(n):
    # np.arange(n, 0, -1)
    vector = np.arange(n, 0, -1) ** 2
    return vector

# Step 2: Example calls
result1 = create_square_vector(5)
result2 = create_square_vector(10)

# Step 3: Display outputs
print("For input 5:", result1)
print("For input 10:", result2)
```

Problem 06:
Generate the following block matrix:

$$
\begin{bmatrix}
1 & 0 & 0 & 2 & 2 \\
0 & 1 & 0 & 2 & 2 \\
0 & 0 & 1 & 2 & 2 \\
25 & 16 & 9 & 4 & 1 \\
\end{bmatrix}
$$

You can use any built-in function to generate the matrix.


```python
import numpy as np

# Step 1: Create 3x3 identity matrix
I = np.eye(3)

# Step 2: Create 3x2 block of 2s
twos_block = np.full((3, 2), 2)

# Step 3: Concatenate these two blocks horizontally
top_block = np.hstack((I, twos_block))

# Step 4: Create the bottom row
bottom_row = np.array([[25, 16, 9, 4, 1]])

# Step 5: Combine top and bottom blocks vertically
final_matrix = np.vstack((top_block, bottom_row))

# Step 6: Display final matrix
print(final_matrix)
```

**Problem 07:**
Plot $y = 2x^2 + 3x + 1$ for $x \in [-3, 3]$ with interval 0.25. You can use any built-in function to plot the graph.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the x range (from -3 to 3 with step 0.25)
x = np.arange(-3, 3.25, 0.25)

# Step 2: Compute y = 2x^2 + 3x + 1
y = 2 * x**2 + 3 * x + 1

# Step 3: Plot the graph
plt.plot(x, y, marker='o', color='blue', linestyle='-', label='y = 2x² + 3x + 1')

# Step 4: Add labels, title, and grid
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = 2x² + 3x + 1')
plt.grid(True)
plt.legend()
plt.show()
```

Problem 08:
Plot  x+y+z=1  for  x,y∈[−3,3]  with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Define x and y range
x = np.arange(-3, 3.5, 0.5)
y = np.arange(-3, 3.5, 0.5)

# Step 2: Create meshgrid for x and y
X, Y = np.meshgrid(x, y)

# Step 3: Calculate Z from equation x + y + z = 1 → z = 1 - x - y
Z = 1 - X - Y

# Step 4: Plot the surface
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k')

# Step 5: Add labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Surface Plot of x + y + z = 1')

plt.show()
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
import numpy as np
import matplotlib.pyplot as plt

# Interval
x = np.arange(-1, 1.01, 0.01)

# Original functions
f_exact = np.exp(2*x) * np.exp(3*x)  # e^(5x)

# Taylor approximations
T2_e2 = 1 + 2*x + 2*x**2
T2_e3 = 1 + 3*x + 4.5*x**2
method1 = T2_e2 * T2_e3  # multiply then truncate higher-order ignored in theory
method2 = 1 + 5*x + 12.5*x**2  # e^(5x)

# Mean Squared Error
mse1 = np.mean((f_exact - method1)**2)
mse2 = np.mean((f_exact - method2)**2)

print("MSE Method 1:", mse1)
print("MSE Method 2:", mse2)

# Plot
plt.figure(figsize=(10,6))
plt.plot(x, f_exact, label='Exact e^(5x)', color='black')
plt.plot(x, method1, '--', label='Method 1: T2[e2]*T2[e3]', color='red')
plt.plot(x, method2, ':', label='Method 2: T2[e5]', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('2nd-order Taylor Approximation Comparison')
plt.legend()
plt.grid(True)
plt.show()
```
